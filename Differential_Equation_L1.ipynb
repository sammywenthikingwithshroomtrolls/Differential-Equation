```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define the Transition Matrix A
# Columns represent 'state today' and rows represent 'state tomorrow'
# (Make sure each column sums to 1.0)
A = np.array([
    [0.5,  0.5,  0.25], # Probability of state 1 tomorrow
    [0.25, 0.0,  0.25], # Probability of state 2 tomorrow
    [0.25, 0.5,  0.5 ]  # Probability of state 3 tomorrow
])

# 2. Define the initial weather state (Day 0)
# e.g., 100% chance it is Sunny (State 1) today
x_today = np.array([1, 0, 0])

# 3. Setup the simulation for 50 days
days = 50
weather_history = np.zeros((days, 3))
weather_history[0, :] = x_today

# 4. Run the discrete system: x_{k+1} = A @ x_k
for i in range(1, days):
    x_tomorrow = A @ x_today
    weather_history[i, :] = x_tomorrow
    x_today = x_tomorrow  # Update today's weather for the next iteration

# 5. Visualize the predictions over time
plt.figure(figsize=(10, 6))
plt.plot(weather_history[:, 0], label="Sunny", color="orange", lw=2)
plt.plot(weather_history[:, 1], label="Cloudy", color="blue", lw=2)
plt.plot(weather_history[:, 2], label="Rainy", color="green", lw=2)

# Customizing the graph
plt.title("Weather Prediction Convergence (Markov Chain)", fontsize=14)
plt.xlabel("Days ($k$)", fontsize=12)
plt.ylabel("Probability", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend(fontsize=12)

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define Transition Matrix A
A = np.array([
    [0.5,  0.5,  0.25],
    [0.25, 0.0,  0.25],
    [0.25, 0.5,  0.5 ]
])

# 2. Initialize xtoday (Column vector: 100% chance of State 1)
xtoday = np.array([[1], [0], [0]])

# 3. Create a matrix to store the results for 50 days (3 states x 50 days)
the_weather = np.zeros((3, 50))
the_weather[:, [0]] = xtoday  # Store Day 0 initial state

# 4. Run the loop updating the values continuously
for k in range(1, 50):
    xtomorrow = A @ xtoday
    the_weather[:, [k]] = xtomorrow  # Record the new day's probabilities
    xtoday = xtomorrow               # Update: Tomorrow becomes today

# 5. Plotting the results
plt.figure(figsize=(10, 5))

# Plot each row of the_weather matrix (State 1, State 2, State 3) over 50 days
plt.plot(the_weather[0, :], label="State 1 (e.g., Sunny)", marker='o', markersize=4)
plt.plot(the_weather[1, :], label="State 2 (e.g., Cloudy)", marker='s', markersize=4)
plt.plot(the_weather[2, :], label="State 3 (e.g., Rainy)", marker='^', markersize=4)

# Graph formatting
plt.title("Weather Probability Convergence Over 50 Days", fontsize=14)
plt.xlabel("Days ($k$)", fontsize=12)
plt.ylabel("Probability", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=11)
plt.ylim(-0.05, 1.05)

# Display the final graph
plt.show()
```

```python
import numpy as np

# 1. Define the Transition Matrix A
# This dictates the probabilities of moving between weather states
A = np.array([
    [0.5,  0.5,  0.25],
    [0.25, 0.0,  0.25],
    [0.25, 0.5,  0.5 ]
])

# 2. Set the initial state for Day 0 (e.g., 100% chance of State 1)
# Using floats (1.0) ensures accurate decimal calculations
xtoday = np.array([[1.0],
                   [0.0],
                   [0.0]])

# Print the starting mathematical vector
print("=" * 35)
print("  Day 0 (Initial State Vector x_0) ")
print("=" * 35)
print(xtoday)
print("\n" + "Starting the 30-day feedback loop..." + "\n")

# 3. Run the loop continuously for a 30-day month
# Each iteration uses the previous day's calculation to find the next
for day in range(1, 31):
    # Calculate tomorrow's vector using matrix multiplication: x_{k+1} = A @ x_k
    xtomorrow = A @ xtoday

    # Print the resulting mathematical vector cleanly
    print(f"--- Day {day} (Vector x_{day}) ---")
    print(np.round(xtomorrow, 6)) # Rounded to 6 decimal places for clean reading
    print("-" * 30)

    # FEEDBACK STEP: Overwrite today's vector with tomorrow's vector
    # This pushes the timeline forward for the next iteration of the loop
    xtoday = xtomorrow
```
